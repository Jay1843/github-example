# Capstone Project - The Battle of the Neighborhoods

### Introduction: Business Problem 

A company approached the data science team with worries over the current economic situation in our country. Their fears have motivated them to plan an emigration of their major business resources and assets to London, where a large portion of business is conducted. They have requested that we provide them with possible locations for their offices to move. They would prefer to all live within the same borough, and thus it would need to have an active out-of-home and night life (restaurants, bars, outdoor activities... etc) as well as be relatively safe and have access to good schools within the area. The distance to city center is not a concern as most of their meetings are virtual, with only a few being in person. They have requested a submission of no less than five borough possibilities.

### Data

The problem requires that boroughs be assessed, grouped and submitted with motivating reasons for why those boroughs meet the definition of ideal. Therefore the following requirements must be met:

The data must assess the standard of living evidenced by the density of venues that match the description given - i.e. that there be a great deal of pubs, cafes, restaurants etc.
The data must include crime statistics.
The data must include the number of good schools - which could be assessed by the number of independent schools in the area.
It was decided that a simple assessment of London Boroughs, the institutions present in those boroughs, and some basic statistics on crime and schools would be needed to find the best borough.
